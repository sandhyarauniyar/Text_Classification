```python
import os
X = []
Y = []
for catg in os.listdir(r"C:\Users\sandhya\Desktop\Newsgroup"):           
    for doc in os.listdir(r"C:\Users\sandhya\Desktop\Newsgroup/"+catg):    
        with open(r"C:\Users\sandhya\Desktop\Newsgroup/"+catg+'/'+doc) as file: 
            X.append((doc,file.read()))
            Y.append(catg)
# X is the tuple having first element as file name and second as all words in the particular file 
# Y is the list of all classes as output
```

```python
# splitting X and Y for training and testing
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y)
```

```python
# buliding vocabulary all proper or meaningful words with key as words in text and value as their frequency
import re,string
vocab = {}
for i in range(len(X_train)):
    text = X_train[i][1]                  # from tuple X_train taking out text
    text = text.lower()                 # converting text to small letters
    text=re.split(r'\W+',text)           # removing punctuations
    for word in text:
    # dont add stop words or numbers or words less than 2
        if not(word.isalpha()) or word in stop_words or len(word)<=2:
            continue
    # if word not in vocab add the word to vocab and initialize the frequency to 1
        if word not in vocab:
                vocab[word] = 1
    # else increase the frequency of word by 1
        else:
                vocab[word] += 1
#vocab            
```

```python
#Sorting the dictionary on basis of frequency of words in descending order
import operator
sorted_vocab = sorted(vocab.items(), key=operator.itemgetter(1),reverse=True)
feature1=sorted_vocab
```

```python
#Plotting graph on no. of words vs frequency 
#On basis of graph we can decide the number of features we want to take
import matplotlib.pyplot as plt
final_feature=[]
feature2=[]
for i in range(len(feature1)):
    final_feature.append(i)
    feature2.append(feature1[i][1])
plt.plot(final_feature,feature2)
plt.axis([0,8000,1,5000])
plt.grid()
plt.show()   
```

```python
# Taking the top 2000 highest frequency words
final_feature = [feature1[i][0] for i in range(2000)]
#final_feature
```

```python
# making the x_train dataset from the vocab in its usual form
X_train_final = np.zeros([len(X_train),len(final_feature)],int)
for i in range(len(X_train)):                 # itearting over X_train
    text = X_train[i][1]                      # from tuple X_train taking out text
    text = text.lower()                       # converting text to small letters so that there is ease in matching
    text = text.strip()                       # removing white spaces from strt & end
    text=re.split(r'\W+',text)              # removing punctuatios
    for word in text:                       # iterating over each word in text
   # if word is present in feature list then add 1 to the final X_train with index equal to index of the word in feature list
        if word in final_feature:  
                X_train_final[i,final_feature.index(word)] += 1
```

```python
# making the x_test dataset from the vocab in its usual form
X_test_final = np.zeros((len(X_test),len(final_feature)))
for i in range(len(X_test)):             # itearting over X_train
    text = X_test[i][1]                   # from tuple X_train taking out text
    text = text.lower()                   # converting text to small letters so that there is ease in matching
    text = text.strip()                   # removing white spaces from strt & end
    text=re.split(r'\W+',text)          # removing punctuatios
    for word in text:                   # iterating over each word in text
#if word is present in feature list then add 1 to the final X_test with index equal to index of the word in feature list     
        if word in final_feature:
             X_test_final[i][final_feature.index(word)] += 1
```

```python
# using sklearn multinomial naive bayes for predicting output classes
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB() 
clf.fit(X_train_final,Y_train)
Y_pred = clf.predict(X_test_final)
```

```python
from sklearn.metrics import classification_report ,confusion_matrix
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

```python
from sklearn.metrics import accuracy_score
print("Score on training data using sklearn :" , clf.score(X_train_final,Y_train))
print("Score on testing data using sklearn :" , clf.score(X_test_final,Y_test))
```

```python
def probability(dictionary,X,current_class):
    output = 0
 # getting the number of words having class as current class
    num_features = len(dictionary[current_class].keys())
        
    for j in range(num_features):   # iterating over each class in  output
    
    # count_of_current_class_with_value_xj is the number of words with class as current class
        count_of_current_class_with_value_xj = dictionary[current_class][final_feature[j]]   + 1
    # count_of_current_class is the total number of words in current class
        count_of_current_class = dictionary[current_class]["total"] +len(dictionary[current_class].keys())-1                       
     # probability_of_xj_for_given_class is the probability the a word belongs to current class
        probability_of_xj_for_given_class = np.log(count_of_current_class_with_value_xj) - (np.log(count_of_current_class))
    # total probability till the current class
        output = output + probability_of_xj_for_given_class
    return output    
```

```python
def predictSinglePoint(dictionary,x):
    classes = dictionary.keys()       # classes has all the possible output classes 
    p_best_class = -1                 # assumption
    best_class = -1000                # assumption
    first_run = True                  # assumption
    for current_class in classes:       # iterating over each class
        if(current_class == "total"):           
            continue
        p_current_class = probability(dictionary,x,current_class)
        if(first_run or p_current_class > best_p):           
            best_p = p_current_class
            best_class = current_class
        first_run = False  
    return best_class    
```

```python
def predict(dictionary,X_test):
    y_pred = []
    for x in X_test:                # iterating over X_test 
        x_class =  predictSinglePoint(dictionary,x)       # here x is a row in X_test i.e X_test[ith row]
        y_pred.append(x_class)                            # y_pred denotes the output list
    return y_pred
```

```python
def fit(X,Y):
    Y = np.array(Y)
    classes = set(Y)                     # getting all the possible classes in output
    count = {} 
    for current_class in classes:          # iterating over each class
        temp = 0
        X_required = X[Y == current_class]       # all rows in X where class is current class

        count[current_class] = {}
        for word in final_feature:               # iterating over all words in feature list
           # count[current_class][word] has all words having class as current class
            count[current_class][word] = (X_required[:,final_feature.index(word)]).sum()
    
         # temp has all words having class as current class
        temp+= count[current_class][word]
        count[current_class]["total"] =  temp
        
   # print(final_feature)    
    return count    
```

```python
def score(Y_test,Y_pred):
    count = 0
    for i in range(len(Y_pred)):
        if Y_test[i] == Y_pred[i]:
            count+=1
    ans = count/len(Y_pred)
    return ans
```
